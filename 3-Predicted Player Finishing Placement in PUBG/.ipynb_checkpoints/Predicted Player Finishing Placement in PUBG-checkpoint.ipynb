## 3. Data Preprocessing

### i. Handling the Null values/Outliers

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
df
```

```python
df['matchType'].value_counts()
```

```python
df.drop(['Unnamed: 0', 'Id', 'groupId', 'matchId'], axis=1, inplace=True)
```

```python
df
```

```python
num_cols = df.select_dtypes('number')
num_cols
```

```python
for col in num_cols:
    plt.figure(figsize=(8,4))
    sns.boxplot(x=col, data=df)
    plt.show()
```

```python
# Define your numerical columns

# Drop outliers using IQR for each column
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
df
```

```python
num_cols = df.select_dtypes('number')
num_cols
```

```python
### ii. Scalling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled = scaler.fit_transform(num_cols)
df_scaled
```

```python
df_scaled = pd.DataFrame(df_scaled, columns = num_cols.columns)
df_scaled
```

```python
# Modifying matchType Column
def get_team_size(x):
    if 'solo' in x:
        return 1
    elif 'duo' in x:
        return 2
    else:
        return 4  # treat squad and unknowns as 4

df['teamSize'] = df['matchType'].apply(get_team_size)
```

```python
df['isFPP'] = df['matchType'].apply(lambda x: 1 if 'fpp' in x else 0)
```

```python
df.drop('matchType', axis=1, inplace=True)
```

```python
df
```

```python
# Concatenate teamSize and isFPP with df_scaled
df_scaled = pd.concat([df_scaled, df[['teamSize', 'isFPP']].reset_index(drop=True)], axis=1)
df_scaled
```

## 4. Feature Selection

```python

# Correlation matrix
corr_matrix = df_scaled.corr().abs()

# Target column correlation
target_corr = corr_matrix['winPlacePerc']

# Drop target from feature-feature matrix to avoid confusion
corr_matrix_features = corr_matrix.drop('winPlacePerc').drop('winPlacePerc', axis=1)

# Upper triangle to avoid duplicates
upper = corr_matrix_features.where(
    np.triu(np.ones(corr_matrix_features.shape), k=1).astype(bool)
)

# Threshold for high correlation
threshold = 0.9

# Print highly correlated feature pairs with their correlation to the target
print("Highly Correlated Feature Pairs (Correlation > 0.9):\n")
for col1 in upper.columns:
    for col2 in upper.columns:
        if upper.loc[col1, col2] > threshold:
            print(f"{col1} and {col2} => Corr: {upper.loc[col1, col2]:.2f} | "
                  f"{col1} vs winPlacePerc: {target_corr[col1]:.2f}, {col2} vs winPlacePerc: {target_corr[col2]:.2f}")
            print('-------------------------------------------------------------------------------------------------')
```

```python
df_scaled.drop(['killPoints', 'maxPlace', 'rankPoints', 'winPoints'], axis=1, inplace=True)
df_scaled
```

## 5. Feature Extraction

```python
X = df_scaled.drop('winPlacePerc', axis=1)
X
```

```python
y = df_scaled[['winPlacePerc']]
y.head()
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pc = pca.fit_transform(X)
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number Component')
plt.ylabel('Variance in %')
plt.title('Explained Variance')
plt.show()
```

```python
np.where(np.cumsum(pca.explained_variance_ratio_) <= 0.90)
```

```python
pca = PCA(n_components=6)
new_data = pca.fit_transform(X)
new_data
```

```python
# column names
columns = [f'Principle Component {i}' for i in range(1,7)]
# creating df
principle_df = pd.DataFrame(data=new_data, columns = columns)
principle_df
```

```python
principle_df = pd.concat([principle_df, y.reset_index(drop=True)], axis=1)
principle_df
```

## 7. Split the data

```python
X = principle_df.drop('winPlacePerc', axis=1)
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

## 8. Fit the data to the model

```python
from sklearn.neighbors import KNeighborsRegressor
model = KNeighborsRegressor(n_neighbors=5)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score
```

## 10. Model Evaluation

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# Predict on training and test sets
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Define a function to print metrics
def evaluate(y_true, y_pred, dataset_name):
    print(f"Evaluation on {dataset_name}:")
    print("R2 Score:", r2_score(y_true, y_pred))
    print("MAE:", mean_absolute_error(y_true, y_pred))
    print("MSE:", mean_squared_error(y_true, y_pred))
    print("RMSE:", np.sqrt(mean_squared_error(y_true, y_pred)))
    print('-' * 40)

# Evaluate on training and test sets
evaluate(y_train, y_train_pred, "Training Set")
evaluate(y_test, y_test_pred, "Test Set")
```
